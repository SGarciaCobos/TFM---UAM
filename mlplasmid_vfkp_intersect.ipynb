## COMBINING RESULTS from mlplasmid AND abricate software - Virulome location

This script allows the combination of results from mlplasmid software (contig prediction as plasmid or chromosome) and abricate software (presence/absence of genes of interest) results. In this case we are using abricate results based on a local virulence genes database. 

```python
import pandas as pd
import os
import json

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)
```

```python
mlplasmid = 'mlplasmids_all_curated.csv'
abricate = 'vfkp_results_cov80id90.csv'
```

```python
# check separator of csv files, in TextEditor for example
dfml = pd.read_csv(mlplasmid, sep=',')
dfab = pd.read_csv(abricate, sep=',')
```

```python
dfml.head()
```

```python
#dfml = dfml[['Isolate_contig', 'Prediction']]
dfml = dfml[['Isolate_contig', 'Prob_Chromosome', 'Prob_Plasmid', 'Prediction']]
```

```python
dfml.head()
```

```python
dfab.head()
```

```python
df = pd.merge(dfab, dfml, how='left', on=['Isolate_contig'])
```

```python
df.head()
```

```python
df.to_csv('mlplasmids_virulenceKp_intersect.csv', index=False)
```

```python
df['sample'] = df.Isolate_contig.str.split('_').str[0]
```

```python
#to add contig column
df['contig'] = df.Isolate_contig.str.split('_').str[1] + df.Isolate_contig.str.split('_').str[2]
```

```python
df.head()
```

```python
df_summary3 = df.groupby(['sample', 'Prob_Plasmid', 'Prediction', 'contig'])['GENE'].apply(list).reset_index(name='genes')
```

```python
df_summary3.to_csv('summaryByContigProb_ml_virulenceKp.csv', index=False)
```
